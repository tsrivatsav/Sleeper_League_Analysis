This notebook serves to extract fantasy football player rankings for dynasty/Superflex PPR leagues from this website: 
https://www.fantraxhq.com/dynasty-football-rankings-superflex-2023/

```python
import pandas as pd
import numpy as np
from requests import get
from bs4 import BeautifulSoup
import re
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
           'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
           'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
           'Accept-Encoding': 'none',
           'Accept-Language': 'en-US,en;q=0.8',
           'Connection': 'keep-alive'}

data_url = 'https://www.fantraxhq.com/dynasty-football-rankings-superflex-2023/'

request = get(data_url, headers=headers)
soup_get = BeautifulSoup(request.content, 'html.parser')
```

```python
def dynasty_table(soupy_object):
    try:
        tbl = soupy_object.find_all('table', attrs = {'id':re.compile('tablepress-2806')})[0].find_all('td')
        table = []
        row = []
        for i in range(len(tbl)):
            row.append(tbl[i].text)
            if i % 7 == 6:
                table.append(row)
                row = []
        dt = pd.DataFrame(table, columns=['Rank', 'Pos Rk', 'Player', 'Pos', 'Team', 'Comments', 'Draft Pick Value'])
    except:
        dt = None
    return dt

dynasty_df = dynasty_table(soup_get)
print(dynasty_df)
```

```python
dynasty_df.to_csv('Datasets/Dynasty_Superflex_Rankings_2023_Oct.csv')
```
