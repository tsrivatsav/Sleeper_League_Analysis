```python
import pandas as pd
import numpy as np
```

```python
player_info_df = pd.read_csv('Datasets/player_info.csv').drop(columns=['Unnamed: 0'])
teams = {}
for owner_id in player_info_df['Owner Id'].unique():
    teams[owner_id] = player_info_df.loc[player_info_df['Owner Id'] == owner_id]
```

```python
for team in teams.values():
    print('----------------')
    print(team.head())
    print('total value', team['Value'].sum())
    team['Value'].hist(legend=True)
```
